```python
dataset = df.to_records(index=False)
```

```python
# Objective function
def fitness_function(individual):
    individual = np.array(individual)
    c = 1
    ts = dataset[dataset['class'] == b'sitting']
    
    if len(ts) == 0:
        return 0  # Return 0 fitness if no instances with class 'sitting' are found
    
    ts = np.array(ts.tolist())
    similarity = np.sum(np.cos(individual.astype(float), ts.astype(float)))

    other_classes = np.unique(dataset['class'])
    other_classes = other_classes[other_classes != b'sitting']

    total_similarity = 0
    for cls in other_classes:
        other_ts = dataset[dataset['class'] == cls]
        other_ts = np.array(other_ts.tolist())
        similarity_i = np.sum(np.cos(individual.astype(float), other_ts.astype(float)))
        total_similarity += similarity_i

    fitness = similarity - c * total_similarity
    return fitness
```

```python
from numpy.random import randint
from numpy.random import rand



# Tournament selection
def selection(pop, scores, k=3):
    selection_ix = randint(len(pop))
    for ix in randint(0, len(pop), k-1):
        if scores[ix] < scores[selection_ix]:
            selection_ix = ix
    return pop[selection_ix]

# Crossover two parents to create two children
def crossover(p1, p2, r_cross):
    c1, c2 = p1.copy(), p2.copy()
    if rand() < r_cross:
        pt = randint(1, len(p1)-2)
        c1 = p1[:pt] + p2[pt:]
        c2 = p2[:pt] + p1[pt:]
    return [c1, c2]

# Mutation operator
def mutation(bitstring, r_mut):
    for i in range(len(bitstring)):
        if rand() < r_mut:
            bitstring[i] = 1 - bitstring[i]

# Genetic algorithm
def genetic_algorithm(fitness_func, n_bits, n_iter, n_pop, r_cross, r_mut):
    pop = [randint(0, 2, n_bits).tolist() for _ in range(n_pop)]
    best, best_fitness = None, float('inf')

    for gen in range(n_iter):
        scores = [fitness_func(c) for c in pop]

        for i in range(n_pop):
            if scores[i] < best_fitness:
                best, best_fitness = pop[i], scores[i]
                print(">%d, new best f(%s) = %.3f" % (gen, pop[i], scores[i]))

        selected = [selection(pop, scores) for _ in range(n_pop)]
        children = []

        for i in range(0, n_pop, 2):
            p1, p2 = selected[i], selected[i+1]
            for c in crossover(p1, p2, r_cross):
                mutation(c, r_mut)
                children.append(c)

        pop = children

        if gen % 10 == 0:
            print("Generation:", gen)

        # Κριτήριο τερματισμού
        # Έλεγχος αν το καλύτερο άτομο πάψει να βελτιώνεται για έναν προκαθορισμένο αριθμό γενεών
        if gen >= 50 and scores[i] == best_fitness:
            print("Termination criteria: Best individual has not improved for 50 generations.")
            break

        # Έλεγχος αν η βελτίωση είναι κάτω από ένα ποσοστό
        if scores[i] < 0.01:
            print("Termination criteria: Improvement is below 1%.")
            break

        # Έλεγχος αν έχουν ξεπεραστεί οι προκαθορισμένες γενεές
        if gen >= 1000:
            print("Termination criteria: Maximum number of generations reached.")
            break

    best_fitness_values = []

    for gen in range(n_iter):
        scores = [fitness_func(c) for c in pop]

        # Υπολογισμός της καλύτερης λύσης στην τρέχουσα γενιά
        best_fitness = float('inf')
        for i in range(n_pop):
            if scores[i] < best_fitness:
                best_fitness = scores[i]

        # Αποθήκευση της καλύτερης λύσης στη λίστα
        best_fitness_values.append(best_fitness)

        # Ο υπόλοιπος κώδικας του γενετικού αλγορίθμου

    return best_fitness_values
# Υπολογισμός της μέσης τιμής των καλύτερων λύσεων για κάθε γενιά
mean_best_fitness = np.mean(best_fitness_values)

# Υπολογισμός του μέσου αριθμού γενεών
mean_num_generations = len(best_fitness_values)

print("Μέση τιμή βέλτιστου:", mean_best_fitness)
print("Μέσος αριθμός γενεών:", mean_num_generations)

# Define the total iterations
n_iter = 100
# Bits
n_bits = len(dataset[0]) - 1
# Define the population size
n_pop = 20
# Crossover rate
r_cross = 0.6
# Mutation rate
r_mut = 0.00 / float(n_bits)

best_fitness_values = genetic_algorithm(fitness_function, n_bits, n_iter, n_pop, r_cross, r_mut)
print('Done!')

# Υπολογισμός της μέσης τιμής των καλύτερων λύσεων για κάθε γενιά
mean_best_fitness_values = np.mean(best_fitness_values)

# Σχεδίαση της καμπύλης εξέλιξης
plt.plot(range(n_iter), best_fitness_values, label='Best Fitness')
plt.axhline(y=mean_best_fitness_values, color='r', linestyle='--', label='Mean Best Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.legend()
plt.show()
```
